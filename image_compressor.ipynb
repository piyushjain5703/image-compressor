```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('my_dp.jpg') #image path wrt to ipynb file
```

```python

img = img[100:500,:,:]
plt.imshow(img)
```

```python
img_size = img.shape
img_size
```

```python
X = img.reshape(img_size[0] * img_size[1], img_size[2])
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=30,init='k-means++',max_iter=100)
model.fit(X)
```

```python
model.cluster_centers_.shape
```

```python
model.labels_
```

```python
X_new = model.cluster_centers_[model.labels_]
X_new = np.reshape(X_new,(img_size[0],img_size[1],img_size[2])) 
X_new = np.clip(X_new.astype('uint8'),0,255)
```

```python
plt.imshow(X_new)
```

```python
plt.imsave('compressed_dp.jpg',X_new)
```
